```python
import glob
import os
import csv
import pandas as pd
from sklearn.metrics import cohen_kappa_score
k=cohen_kappa_score
```

```python
df0 = pd.read_csv (r'nyt_stance-2/nyt_stance_0.csv') #nyt_stance_0
df0["label"]=""  #Adding a column for final label

df0['annotation_8'].fillna('missing', inplace=True) #Handling empty cells in the data file
df0['annotation_9'].fillna('missing', inplace=True)
df0['annotation_10'].fillna('missing', inplace=True)
df0['annotation_7'].fillna('missing', inplace=True)

x8=k(df0.annotation_8, df0.annotation_9) #annotation_8
print (x8)
y8=k(df0.annotation_8, df0.annotation_10)
print (y8)
z8=k(df0.annotation_8, df0.annotation_7)
print (z8)
avg8=(x8+y8+z8)/3
print('\nAverage Kappa score for annotation_8 is ', avg8)
print('\n')

x9=k(df0.annotation_9, df0.annotation_8) #annotation_9
print (x9)
y9=k(df0.annotation_9, df0.annotation_10)
print (y9)
z9=k(df0.annotation_9, df0.annotation_7)
print (z9)
avg9=(x9+y9+z9)/3
print('\nAverage Kappa score for annotation_9 is ', avg9)
print('\n')

x10=k(df0.annotation_10, df0.annotation_8) #annotation_10
print (x10)
y10=k(df0.annotation_10, df0.annotation_9)
print (y10)
z10=k(df0.annotation_10, df0.annotation_7)
print (z10)
avg10=(x10+y10+z10)/3
print('\nAverage Kappa score for annotation_10 is ', avg10)
print('\n')

x7=k(df0.annotation_7, df0.annotation_8) #annotation_7
print (x7)
y7=k(df0.annotation_7, df0.annotation_9)
print (y7)
z7=k(df0.annotation_7, df0.annotation_10)
print (z7)
avg7=(x7+y7+z7)/3
print('\nAverage Kappa score for annotation_7 is ', avg7)
print('\n')
```

```python
for i in df0.index:
    pm=0
    am=0
    uc=0
    
    if avg8>=0.2: #Checking if annotator is reliable
        if df0.annotation_8[i]=='pro-mitigation':
            pm=pm+1
        elif df0.annotation_8[i]=='anti-mitigation':
            am=am+1
        elif df0.annotation_8[i]=='unclear':
            uc=uc+1
    
    if avg9>=0.2: #Checking if annotator is reliable    
        if df0.annotation_9[i]=='pro-mitigation':
            pm=pm+1
        elif df0.annotation_9[i]=='anti-mitigation':
            am=am+1
        elif df0.annotation_9[i]=='unclear':
            uc=uc+1
    
    if avg10>=0.2: #Checking if annotator is reliable
        if df0.annotation_10[i]=='pro-mitigation':
            pm=pm+1
        elif df0.annotation_10[i]=='anti-mitigation':
            am=am+1
        elif df0.annotation_10[i]=='unclear':
            uc=uc+1
    
    if avg7>=0.2: #Checking if annotator is reliable
        if df0.annotation_7[i]=='pro-mitigation':
            pm=pm+1
        elif df0.annotation_7[i]=='anti-mitigation':
            am=am+1
        elif df0.annotation_7[i]=='unclear':
            uc=uc+1
    
    if pm>am and pm>uc:
        df0.label[i]="pro-mitigation"
    elif am>pm and am>uc:
        df0.label[i]="anti-mitigation"
    elif uc>pm and uc>am:
        df0.label[i]="unclear"
    else:
        df0.label[i]=df0.annotation_8[i] #In case of ties

df0=df0.drop(columns=df0.columns.values.tolist()[1:len(df0.columns.values.tolist())-1])
df0.to_csv("nyt_st_0.csv", mode='a', index=False)
```
